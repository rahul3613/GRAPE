```python
import pandas as pd
import utm
```

```python
pd.set_option('display.precision', 10)
```

```python
f_name = "validationData"
```

```python
df = pd.read_csv("uji/raw_data/UJIndoorLoc/data/{}.csv".format(f_name))
```

```python
df[["LONGITUDE", "LATITUDE"]] = df[["LONGITUDE", "LATITUDE"]]/100000
df
```

```python
df[['x', 'y']] = df.apply(lambda row: pd.Series(utm.from_latlon(row['LATITUDE'], row['LONGITUDE'])[:2]), axis=1)
df
```

```python
df[['x_scaled']] = df[['x']] - df[['x']].mean()
df[['y_scaled']] = df[['y']] - df[['y']].mean()
df
```

```python
df.to_csv("{}.csv".format(f_name))
```
